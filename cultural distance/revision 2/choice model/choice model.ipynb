# **filter for neighbourhood**

```
import numpy as np
import pandas as pd
import feather
df = feather.read_dataframe("/content/drive/My Drive/Dissertation/airbnb/cultural distance/choice model/data.feather")
#nbh = feather.read_dataframe("/content/drive/My Drive/Dissertation/airbnb/cultural distance/choice model/nbh.feather")
#nbh = nbh[['id', 'neighbourhood_cleansed']]
#nbh = nbh.drop_duplicates(subset=['id'])
#df = pd.merge(left=df, right=nbh, left_on='listing_id', right_on='id', how='left')
#feather.write_dataframe(df, "/content/drive/My Drive/Dissertation/airbnb/cultural distance/choice model/data.feather")
```

```
df = df[df.neighbourhood_cleansed == "Lido"]
df.reset_index(drop=True, inplace=True)
x = df.listing_id.unique().tolist()
x = ["{}{}".format('sh_', i) for i in x]
var = ['listing_id', 'review_id', 'date', 'reviewer_id', 'country', 'cult_dst_6', 'id_date', 'yyyymm', 'period', 'neighbourhood_cleansed']
var = var + x
df = df[df.columns.intersection(var)]
feather.write_dataframe(df, "/content/drive/My Drive/Dissertation/airbnb/cultural distance/choice model/san marco/data.feather")
```

# **format to stata**

```
%%R
x = as.data.frame(df)
y <- as.data.frame(x$idx)
z <- cbind(x, y)
z$listing_id <- ifelse(z$listing_id==TRUE, 1, 0)
z$idx <- NULL
z$sh_x_dst <- z$sh*z$cult_dst_6
write.dta(z, "/content/drive/My Drive/Dissertation/airbnb/cultural distance/choice model/murano/data.dta")
#export z to .dat
```
